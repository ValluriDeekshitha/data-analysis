```python
## construction of simple sample graph
x=[1,2,3,4,5]
y=[10,30,50,70,90]
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
z=[12.5,34.5,23,55.3,64.5]
plt.plot(x,y,label="first")
plt.plot(x,z,label="second")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title("Second Graph")
plt.show()
```

```python
li=[1,2,3,4,5,6]
li1=[10.5,11.5,13.5,15.6,12.5,18.9]
plt.plot(li,li1)
plt.axis([0,10,0,25]) # axis range
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Sample-3")
plt.show()
```

```python
## ggplot
from matplotlib import style
style.use('ggplot')
days=[1,2,3,4,5]
temperature=[36.1,34.6,45.2,39.9,45.8]
plt.xlabel("days",fontsize=15)
plt.xlabel("temperature",fontsize=10)
plt.plot(days,temperature,color='g',marker='o',linewidth=12)
```

```python
vijaya_temp=[33.3,40.5,28.7,26.9,31.3]
plt.plot(days,vijaya_temp,'m*:',linewidth=5,label='vijaya')
plt.plot(days,temperature,color='g',marker='o',linewidth=3,label='tnk')
plt.title('tnk-temp Vs vijaya-temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.legend(loc=3)
plt.grid(color="r",linestyle="-",linewidth=2)
plt.show()
```

```python
##used to draw stacked area plot
## displays the complete data for visualisation
## it displays in form of constituents
# it mostly behaves like pie chart
```

```python
x=[9,3,5,4,8]
y=[10,6,4,5,4]
z=[7,8,9,7,4]
a=np.array([8,4,9,6,2])
plt.stackplot(x,y,z,a)
plt.show()
```

```python
f=[1,2,3,4,5]
s={'y1':[5,6,7,2,9],'y2':[1,3,5,6,7],'y3':[2,4,6,8,7],'y4':[2,4,1,5,7]}
plt.stackplot(f,s.values())
plt.show()
```

```python
plt.stackplot(f,s.values(),labels=s.keys())
plt.legend(loc="upper left")
plt.title("stack plot")
plt.show()
```
